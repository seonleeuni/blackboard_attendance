```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os

# Wirte the path of excel file exists(file 에 excel 파일 하나만 있어야함)
path_dir = 'C:/Users/UNIST/Desktop/1학년 2학기/뇌인행/scoring/list' # <- CHANGE THIS INPUT
file_list = os.listdir(path_dir)
data = pd.read_excel(path_dir + '/' + file_list[0])
```

```python
#Call only last exam score
data = data[data.columns[len(data.columns)-1]]

# value_counts for countplot
draw = pd.DataFrame(data.value_counts())
draw = draw.reset_index()
draw.columns = ['score','count']
```

```python
# Plot & Print stochastic values
font = 20
plt.figure(figsize = (15,10))
sns.lineplot(data = draw, x = 'score', y = 'count')
plt.xticks(fontsize =font)
plt.yticks(range(0,max(draw['count'])+1),fontsize = 20font)
plt.xlabel('score', fontsize = font)
plt.ylabel('count', fontsize = font)
plt.grid(axis = 'y' ,which = 'both')
plt.savefig('fig.png')
print('min: ',min(data),' |max: ', max(data),' |average: ', round(data.mean(),2),' |median: ', data.median())
```
